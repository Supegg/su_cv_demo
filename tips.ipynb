````python
# 打开、显示、关闭图片
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
print("OpenCV's version:", cv.__version__)
````

````python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()    
````

````python
path = 'pics\elegance.png'
img = cv.imread(path)
cv.imshow('elegance',img)
cv.waitKey(0)
cv.destroyAllWindows()
# cv.imwrite('pics\my.jpg', img)
````

````python
img = cv.imread(path)
show(img) 
````
