# 条码检测

* 可一次检出所有码
* 仅支持二维码检测

```python
from utils import show
import cv2 as cv
import numpy as np
import pyboof as pb

img = cv.imread(r'pics\barcode_sample1.png')
show(img)
```

```python
print('Detect with pyboof')

detector = pb.FactoryFiducial(np.uint8).qrcode() # just qr code
img = cv.imread(r'pics\barcode_sample1.png', 0)
boof_img = pb.ndarray_to_boof(img)

detector.detect(boof_img)
for detection in detector.detections:
    # 先进行缩放操作，再取整
    points = [(int(p[0]), int(p[1])) for p in detection.bounds.convert_tuple()] 
    print(points, detection.message)
    
    
```
